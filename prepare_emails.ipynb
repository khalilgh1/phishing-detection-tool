```python
import pandas as pd
```

```python
IN = "emails.csv"
OUT = "emails_cleaned.csv"
df = pd.read_csv(IN)

```

```python
df = df.dropna(subset=["subject"])
df = df.drop(columns=["receiver", "date", "urls"], errors='ignore')
df['text'] = df['sender'].fillna('') + ' ' + df['subject'].fillna('') + ' ' + df['body'].fillna('')
df = df[['text', 'label']]
df.to_csv("emails_cleaned.csv", index=False)
```
